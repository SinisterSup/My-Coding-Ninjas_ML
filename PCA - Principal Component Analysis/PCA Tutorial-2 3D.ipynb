```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA                            
```

```python
np.random.seed(2343243)
```

```python
mean_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1 = np.random.multivariate_normal(mean_vec1,cov_mat1,100)         
```

```python
mean_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2 = np.random.multivariate_normal(mean_vec2,cov_mat2,100)
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D, proj3d

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
ax.plot(class1[:,0], class1[:,1], class1[:,2],'o')
ax.plot(class2[:,0], class2[:,1], class2[:,2],'^')
plt.show()           
```

```python
X_data = np.concatenate((class1, class2))
X_data
```

```python
pca = PCA(n_components=2)
X_transformed = pca.fit_transform(X_data)
X_transformed
```

```python
pca.components_
```

```python
plt.plot(X_transformed[:100,0], X_transformed[:100,1], 'o')
plt.plot(X_transformed[100:200,0], X_transformed[100:200,1], '^')
plt.show()
```

```python
original_X = pca.inverse_transform(X_transformed)
original_X
```

## Trying to Build PCA for Scratch using Knowledge of Math 
1. ### CoVariance    
2. ### Eigen_Vectors 
3. ### Eigen_values

```python
X_data_transpose = X_data.T
covar = np.cov(X_data_transpose)
print(covar.shape)
print(covar)
```

```python
eig_val, eig_vectors = np.linalg.eig(covar)
eig_val, eig_vectors
```

```python
eig_val_vector_pair = []
for i in range(len(eig_val)):
    eig_vec = eig_vectors[:i]
    eig_val_vector_pair.append((eig_val[i], eig_vec))
    
eig_val_vector_pair.sort(reverse=True)
eig_val_vector_pair
```
